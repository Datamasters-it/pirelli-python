## Resampling
##### Downsampling: ricampionamento ad una frequenza inferiore
##### Upsampling: ricampionamento ad una frequenza maggiore

````python
# definizione colonne per cui fare il downsampling
cols = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']

# operazione di resampling (media valori per settimane)

df_media_settimanale = df[cols].resample('W').mean()
df_media_settimanale.tail(5)
````

````python
# visualizzazione dimensione DataFrame dopo resampling
df_media_settimanale.shape
````

````python
# Date di inizio e fine analisi
start, end = '2017-01', '2017-06'

# plot serie temporali giornaliere e settimanali (media)

# creazione grafico
fig, ax = plt.subplots()

# plot serie temporale giornaliera
ax.plot(df.loc[start:end, 'Wind'], marker='.', linestyle='--', linewidth=0.5, label='Giornaliero')

# plot serie temporale media settimanale
ax.plot(df_media_settimanale.loc[start:end, 'Wind'], marker='o', markersize=8, linestyle='-', label='Resampling Media Settimanale')

# impostazione etichetta asse y
ax.set_ylabel('Produzione energia Eolica (GWh)')

# visualizzazione legenda
ax.legend()

plt.show()
````

````python
# tipologie di intervalli temporali da poter utilizzare con resample:
#
# B - giorni lavorativi
# W - settimane
# A - anni
# D - giorni
# H - ore
# T - minuti
# S - secondi
````

````python
# calcolo consumo totale annuale
df_media_annuale = df[cols].resample('A').sum()
# impostazione Anno come indice
````

````python
# visualizzazione indici DataFrame
df_media_annuale.index
# A-DEC Anno Dicembre(finale)
# frequenza annuale
````

````python
# impostazione indice
df_media_annuale = df_media_annuale.set_index(df_media_annuale.index.year)
df_media_annuale.index.name = 'Anno'
df_media_annuale.tail(5)
````

````python
# visualizzazione del rapporto tra l'energia prodotta e quella consumata
df_media_annuale['Wind+Solar/Consumption'] = df_media_annuale['Wind+Solar'] / df_media_annuale['Consumption']

# visualizzazione ultimi cinque anni
df_media_annuale.tail(5)
````

````python
# visualizzazione grafico a barre del rapporto tra l'energia prodotta e quella consumata 
# per anno dal 2012
ax = df_media_annuale.loc[2012:, 'Wind+Solar/Consumption'].plot.bar()
ax.set_ylabel('Rapporto')
ax.set_title('Rapporto tra energia solare e eolica prodotta e quella consumata')
````

## Rolling window


Come funziona rolling

![Title](img/rolling.png)



Il rolling in Pandas Ã¨ una tecnica potente per calcolare statistiche su una finestra mobile di valori. Ãˆ particolarmente utile per analizzare serie temporali e dati di trend. In questo approfondimento, esamineremo il concetto di rolling, come utilizzare il metodo `rolling()` in Pandas, e alcuni esempi pratici.

### Concetti di Base

Il rolling Ã¨ il processo di applicare una funzione statistica a una finestra mobile di dati. Ad esempio, una media mobile di 3 giorni calcolerÃ  la media dei valori per ogni finestra di 3 giorni consecutivi.

### Sintassi di Base

La sintassi di base per l'operazione di rolling in Pandas Ã¨:

```python
df['column'].rolling(window=n).function()
```


Dove:

-   `window=n` specifica la dimensione della finestra mobile.
-   `function()` Ã¨ la funzione statistica applicata alla finestra mobile (ad esempio, `.mean()`, `.sum()`, `.std()`, etc.).

es.

```python
df['Rolling_Mean'] = df['Value'].rolling(window=3).mean()
```


````python
# definizione colonne per cui fare il downsampling
cols = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']

# funzione di rolling (media dei cinque valori nella finestra)
df_rolling = df[cols].rolling(window=5, center=True).mean()

# visualizzazione prime 10 entry DataFrame
df_rolling.head(10)
````

````python
# visualizzazione prime 10 entry dopo rolling
df_rolling.head(10)
````

````python
# Date di inizio e fine analisi
start, end = '2017-02', '2017-07'

# plot serie temporali giornaliere, resample settimanale (media) e media mobile su 5 giorni

# creazione grafico
fig, ax = plt.subplots()

# plot serie temporale giornaliera
ax.plot(df.loc[start:end, 'Wind'], marker='.', linestyle='-', linewidth=0.5, label='Giornaliero')

# plot serie temporale media settimanale
ax.plot(df_media_settimanale.loc[start:end, 'Wind'], marker='o', markersize=8, linestyle='-', label='Resampling media settimanale')

# plot serie temporale media mobile su 5 giorni
ax.plot(df_rolling.loc[start:end, 'Wind'], marker='.', linestyle='-', label='5-d Rolling Mean')

# impostazione etichetta asse y
ax.set_ylabel('Produzione energia eolica (GWh)')

# visualizzazione legenda
ax.legend()

plt.show()
````
