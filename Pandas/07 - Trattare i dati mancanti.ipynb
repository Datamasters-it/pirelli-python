```python
df = pd.read_csv("../dataset/healthcare-dataset-stroke-data.csv")
df.info()
```

```python
random_rows = np.random.permutation(df.values.shape[0])[:500]
# sceglie delle colonne e assegna dei valori NaN a tutte le colonne scelte appartenenti alle righe che fanno parte della permutazione
# random_cols = np.random.permutation(df.values.shape[1])[:3]
# df.iloc[
#    random_rows,
#    random_cols
# ] = np.nan
# df.info()
```

```python
# Per ogni riga della permutazione prende 3 colonne casuali e mette i valori NaN in quelle colonne

for index, row in df.loc[random_rows].iterrows():
    random_cols = np.random.permutation(df.values.shape[1])[:3]
    df.iloc[
        index,
        random_cols
    ] = np.nan
```

```python
df.info()
```

```python
# KNNImputer usa l'algoritmo KNN per sostituire i valori mancanti con la media dei K valori più simili alla riga considerata

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')

imputed_data = imputer.fit_transform(df.select_dtypes(include=[np.number]))  # Only numeric columns

# Convert the array back to a DataFrame
imputed_df = pd.DataFrame(imputed_data, columns=df.select_dtypes(include=[np.number]).columns)

# Replace the numeric columns in the original dataframe with imputed columns
df[imputed_df.columns] = imputed_df
```

```python
df.info()
```

```python
# SimpleImputer usa metodi più semplici per sostituire un valore mancante; si può applicare anche alle stringhe

from sklearn.impute import SimpleImputer

imputer_most_frequent = SimpleImputer(strategy='most_frequent')
non_numeric_columns = df.select_dtypes(exclude=[np.number]).columns
print(non_numeric_columns)

# Impute each non-numeric column
for c in non_numeric_columns:
    # Check if the column is entirely NaN
    if df[c].isnull().all():
        print(f"Column {c} is entirely NaN.")
    else:
        try:
            # Reshape data and impute
            d = imputer_most_frequent.fit_transform(df[[c]])
            df[c] = df[c].replace(df[c].values, d)
        except Exception as e:
            print(f"Error imputing column {c}: {e}")

```
