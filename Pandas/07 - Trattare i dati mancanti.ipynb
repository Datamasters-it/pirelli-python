```python
# Per ogni riga della permutazione prende 3 colonne casuali e mette i valori NaN in quelle colonne
# iterrows mi dà la possibilità di ciclare sulle righe di un dataframe
# in questo caso sto ciclando sulle sole righe casuali estratte in precedenza

for index, row in df.loc[random_rows].iterrows():
    random_cols = np.random.permutation(df.values.shape[1])[:3]
    df.iloc[
        index,
        random_cols
    ] = np.nan
```

```python
df.info()
```

```python
# KNNImputer usa l'algoritmo KNN per sostituire i valori mancanti con la media dei K valori più simili alla riga considerata

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')

imputed_data = imputer.fit_transform(df.select_dtypes(include=[np.number]))  # Only numeric columns

# Convert the array back to a DataFrame
imputed_df = pd.DataFrame(imputed_data, columns=df.select_dtypes(include=[np.number]).columns)

# Replace the numeric columns in the original dataframe with imputed columns
df[imputed_df.columns] = imputed_df
```

```python
df.info()
```

```python
# SimpleImputer usa metodi più semplici per sostituire un valore mancante; si può applicare anche alle stringhe

from sklearn.impute import SimpleImputer

imputer_most_frequent = SimpleImputer(strategy='most_frequent')
non_numeric_columns = df.select_dtypes(exclude=[np.number]).columns
print(non_numeric_columns)

# Impute each non-numeric column
for c in non_numeric_columns:
    # Check if the column is entirely NaN
    if df[c].isnull().all():
        print(f"Column {c} is entirely NaN.")
    else:
        try:
            # Reshape data and impute
            d = imputer_most_frequent.fit_transform(df[[c]])
            df[c] = df[c].replace(df[c].values, d)
        except Exception as e:
            print(f"Error imputing column {c}: {e}")

```

### Esercizio: Rilevazione e Gestione degli Outlier con il Metodo dell'Interquartile Range (IQR)

**Obiettivo**: Imparare a identificare e gestire gli outlier in un dataset utilizzando il metodo dell'Interquartile Range (IQR). Questo esercizio aiuterà a comprendere l'impatto degli outlier sull'analisi dei dati e come mitigarne efficacemente l'effetto.

#### Dataset:

Utilizzare `healthcare-dataset-stroke-data.csv`

#### Attività:

**Attività 1: Preparazione dei Dati**

-   Caricare il dataset in un DataFrame di pandas.
-   Eseguire un\'analisi preliminare: visualizzare le prime righe e recuperare le sole colonne numeriche su cui ha senso parlare di outlier.

**Attività 2: Calcolo dell'IQR**

-   Per ciascuna colonna numerica, calcolare il primo quartile (Q1) e il terzo quartile (Q3).
-   Calcolare l'IQR (Q3 - Q1) per queste colonne.

**Attività 3: Identificazione degli Outlier**

-   Definire gli outlier come i punti sotto `(Q1 - 1.5 * IQR)` o sopra `(Q3 + 1.5 * IQR)`.

**Attività 4: Gestione degli Outlier**

-   Implementare due strategie per gestire gli outlier:
    1.  **Rimozione degli Outlier**: Creare un nuovo DataFrame che esclude gli outlier.
    2.  **Limitazione degli Outlier**: Sostituire gli outlier con i valori di confine (usando il metodo `clip` sulla series
-   Stampare un boxplot prima e dopo la gestione degli outlier

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../dataset/healthcare-dataset-stroke-data.csv")

Q1 = df[['avg_glucose_level']].quantile(0.25) # primo quartile
Q3 = df[['avg_glucose_level']].quantile(0.75) # terzo quartile

print(Q1)
print(Q3)

IQR = Q3 - Q1
print(IQR)

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

within_bounds = (df[['avg_glucose_level']] >= lower_bound) & (df[['avg_glucose_level']] <= upper_bound)

df_without = df.loc[within_bounds.iloc[:,0]]

df_without.shape

# outliers_specific.any(axis=1)

#float_columns = ['age', 'avg_glucose_level', 'bmi']

# print(float_columns)

sns.boxplot(data = df['avg_glucose_level'], orient = "h")
plt.show()
```

```python
sns.boxplot(data = df_without['avg_glucose_level'], orient = "h")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../dataset/healthcare-dataset-stroke-data.csv")

float_columns = ['avg_glucose_level', 'age', 'bmi']


Q1 = df[float_columns].quantile(0.25) # primo quartile
Q3 = df[float_columns].quantile(0.75) # terzo quartile

print(Q1)
print(Q3)

IQR = Q3 - Q1
print(IQR)

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

within_bounds = (df[float_columns] >= lower_bound) & (df[float_columns] <= upper_bound)

df_without = df.loc[within_bounds.iloc[:,0]]

df_without.shape

# outliers_specific.any(axis=1)

#float_columns = ['age', 'avg_glucose_level', 'bmi']

# print(float_columns)

sns.boxplot(data = df[float_columns], orient = "h")
plt.show()
```

```python
sns.boxplot(data = df_without[float_columns], orient = "h")
plt.show()
```

```python
df_capped_specific = df.copy()
for column in float_columns:
    df_capped_specific[column] = df_capped_specific[column].clip(lower=lower_bound[column], upper=upper_bound[column])

sns.boxplot(data = df_capped_specific[float_columns], orient = "h")
plt.show()
```
