## Combinare pi√π DataFrame

```python
# import librerie
import pandas as pd
```

```python
# definizione DataSet
df1 = pd.DataFrame({
"nome":["Mark","Andrea","Luca","Alex","Jack","Max","Lou","Kim","Frank","Sam","Paul"]
,"livello":["5","6","4","4","5","6","3","5","4","6","6"]
,"zona":["Sud","Nord","Sud","Nord","Sud","Sud","Centro","Centro","Nord","Sud","Centro"]
,"incassi":[50000,52000,90000,34000,42000,72000,49000,55000,67000,65000,67000]
,"spese":[42000,43000,50000,44000,38000,39000,42000,60000,39000,44000,45000]})

df2 = pd.DataFrame({
"nome":["George","Nina","Moira","Carl","Bob","Keller","Dimitri","Xavier","Willy","Antoine","David"]
,"livello":["3","5","5","4","5","6","6","5","4","3","6"]
,"zona":["Nord","Centro","Sud","Nord","Centro","Sud","Nord","Centro","Nord","Sud","Sud"]
,"incassi":[60000,42000,80000,24000,23000,62000,39000,45000,73000,51000,78000]
,"spese":[32000,23000,60000,45000,38000,36000,48000,40000,69000,94000,25000]})

df1
```

```python
df2
```

```python
# concatenazione di due DataFrame
df_arr = [df1, df2]

df = pd.concat(df_arr)

df
```

```python
# concatenazione di due DataFrame ignorando gli indici
df_arr = [df1, df2]

df = pd.concat(df_arr, ignore_index=True)

df
```

```python
# concatenazione di DataFrame e Series
s = pd.Series(["23", "25", "37", "28", "41", "58", "55", "47", "29", "30", "31"])

df = pd.concat([df1, s], axis=1)
df
```

```python
# concatenazione di DataFrame e Series ignorando gli indici iniziali
df = pd.concat([s, df1], axis=1, ignore_index=True)
df
```

```python
#aggiungere un DataFrame ad un altro
result = pd.concat([df, df2])

result
```

```python
#aggiungere un DataFrame ad un altro ignorando gli indici iniziali
result = pd.concat([df, df2], ignore_index=True)

result
```

```python
# merge

# definizione DataFrame
df3 =  pd.DataFrame({ 'utente_id': ["ut_0003", "ut_0004", "ut_0005", "ut_0006", "ut_0007"]
,"nome":["Luca","Alex","Jack","Max","Lou"]
,"cognome":["Verdi","Valli","Reacher","Power","Biga"]})

df4 =  pd.DataFrame({ 'utente_id': ["ut_0001", "ut_0002", "ut_0003", "ut_0004", "ut_0005"]
,"et√†":["23","25","37","28","41"]})
```

```python
df3
```

```python
df4
```

```python
# merge
df_merge_df3_df4 = pd.merge(df3, df4)

df_merge_df3_df4
```

```python
# merge in base a attributo comune
df_merge_df3_df4 = pd.merge(df3, df4, on='utente_id')

df_merge_df3_df4
```

```python
# left merge
df_left_merge_df3_df4 = pd.merge(df3, df4, how='left')

df_left_merge_df3_df4
```

```python
# right merge
df_right_merge_df3_df4 = pd.merge(df3, df4, how='right')

df_right_merge_df3_df4
```

```python
# inner merge
df_right_merge_df3_df4 = pd.merge(df3, df4, how='inner')

df_right_merge_df3_df4
```

```python
# outer merge
df_right_merge_df3_df4 = pd.merge(df3, df4, how='outer')

df_right_merge_df3_df4
```
